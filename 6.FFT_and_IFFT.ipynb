## FFT 

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1 D FFT

```python
t=np.arange(128)
sp=np.fft.fft(np.sin(t))
freq=np.fft.fftfreq(t.shape[-1])
print(sp,freq)
```

```python
plt.subplot(2,1,1)
plt.title('Real Part')
plt.plot(freq,sp.real)

plt.subplots_adjust(hspace=0.2)

plt.subplot(2,1,2)
plt.title('Imaginary Part')
plt.plot(freq,sp.imag)

plt.show()
```

## IFFT

```python
t=np.arange(500)
n=np.zeros((500,),dtype=complex)
n[40:80]=np.exp(1j*np.random.uniform(0, 2*np.pi,(40,)))
print(t,n)
```

```python
s=np.fft.ifft(n)
plt.plot(t,s.real)
plt.plot(t,s.imag)
plt.title('IFFT')
plt.show()
```
