```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(5)
print(x)
```

```python
y=x
print(y)
```

```python
plt.plot(x,y,'o--')
plt.plot(x,-y,'o-')
plt.title('y=x and y=-x')
plt.show()
```

```python
N=11
x=np.linspace(0,10,N)
print(x)
y=x
```

```python
plt.plot(x,y,'o--')
plt.show()
```

```python
y=np.logspace(0.1,1,N)
plt.plot(x,y,'o--')
plt.show()
```

```python
y=np.geomspace(0.1,1000,N)
print(y)
plt.plot(x,y,'o--')
plt.show()
```
