# SIN COS TAN

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
low=-np.pi
high=np.pi
N=300
x=np.linspace(low,high,N)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
y=np.tan(x)
plt.plot(x,y)
plt.show()
```

# Hyperbolic Functions

```python
y=np.sinh(x)
plt.plot(x,y)
plt.show()
```

```python
y=np.cosh(x)
plt.plot(x,y)
plt.show()
```

```python
y=np.tanh(x)
plt.plot(x,y)
plt.show()
```

# Exponents

```python
x=np.arange(50)
y=np.exp(x)
plt.plot(x,y)
plt.show()
```

```python
# exp minus 1
y=np.expm1(x)
plt.plot(x,y)
plt.show()
```

```python
# 2 ** y
y=np.exp2(x)
plt.plot(x,y)
plt.show()
```

# Logarithmic Functions

```python
x=np.linspace(1,10,10)
y=np.log(x)
plt.plot(x,y)
plt.show()
```

```python
y=np.log10(x)
plt.plot(x,y)
plt.show()
```

```python
y=np.log2(x)
plt.plot(x,y)
plt.show()
```

# CONVOLUTION

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(np.convolve(a,b))
```
